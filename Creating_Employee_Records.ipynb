### Criando banco de dados

```python
conect = sqlite3.connect('register.db')
menseng = conect.cursor()
menseng.execute('''CREATE TABLE employ (
    ide INTEGER NOT NULL,
    name TEXT NOT NULL,
    email TEXT NOT NULL,
    department TEXT NOT NULL,
    salary FLOAT NOT NULL,
    birth_date TEXT NOT NULL
    )
''')

conect.commit()

conect.close()
```
